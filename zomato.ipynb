```python
# immporting libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


# create data frame 

dataframe = pd.read_csv("Zomato data .csv")
print(dataframe)


# data cleaning

#  convert the data type of column-rate

def handlerate(value):
    value = str(value).split('/')
    value = value[0];
    return float(value)

dataframe['rate'] = dataframe['rate'].apply(handlerate)
print(dataframe.head())


# check is there null data or not

print(dataframe.info())


# TASKS:--

# Ques - 1 --> What type of customers do the majority of customers order from?

dataframe.head()
sns.countplot(x=dataframe['listed_in(type)'])
plt.xlabel("type of restaurant")


#  Ans - 1 --> Conclusion -- major no. of restaurant falls in dining category



# Ques - 2 --> How many votes has each restaurant received from customers ?


grouped_data = dataframe.groupby('listed_in(type)')['votes'].sum()
result = pd.DataFrame({'votes':grouped_data})
plt.plot(result,c="green",marker = "o")
plt.xlabel("Type of Restaurant",c="red",size =20)
plt.ylabel("votes",c="red",size=20)

# Ans - 2 --> Dinning restaurants has received maximum votes




# Ques - 3 --> what are the ratings that majority of restaurants have received?

plt.hist(dataframe['rate'],bins=5)
plt.title("ratings distribution")
plt.show()


# Ans - 3 --> The majority restaurants received ratings from 3.5 to 4




# Ques - 4 --> Zomato has observed that  most couples order most of the their food online . what is their average spending on each order?



couple_data = dataframe['approx_cost(for two people)']
sns.countplot(x=couple_data)


# Ans - 4 --> The majority of couples prefer restaurants with an approximate cost of 300rs.




# Ques - 5 --> which mode (online or offline) has received the maximum rating ?



plt.figure(figsize=(6,6))
sns.boxplot(x='online_order',y='rate',data=dataframe)

# Ans - 5 --> offline order received lower rating in comparison to online order 
```
